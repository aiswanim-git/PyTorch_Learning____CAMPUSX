**Comparing performance difference between GPU and CPU**

```python
import time

size=10000
matrix1_cpu=torch.randn(size,size)
matrix2_cpu=torch.randn(size,size)

#measure time on CPU
start_time=time.time()
result_cpu=torch.matmul(matrix1_cpu,matrix2_cpu)
cpu_time=time.time()-start_time
print(f"Time taken by CPU: {cpu_time} seconds")

#move matrices to GPU
matrix1_gpu=matrix1_cpu.to('cuda')
matrix2_gpu=matrix2_cpu.to('cuda')

#measure time on GPU
start_time=time.time()
result_gpu=torch.matmul(matrix1_gpu,matrix2_gpu)
gpu_time=time.time()-start_time
print(f"Time taken by GPU: {gpu_time} seconds")

#compare time taken
print(f"GPU is {cpu_time/gpu_time} times faster than CPU")
```

```python
import torch

#inputs
x=torch.tensor(6.7) #input feature
y=torch.tensor(0.0) #true label(binary)

w=torch.tensor(1.0) #weight
b=torch.tensor(0.0) #bias
```

```python
def binary_cross_entropy_loss(prediction,target):
  epsilon=1e-8 #to prevent log(0)
  prediction=torch.clamp(prediction,epsilon,1-epsilon)
  loss=-(target*torch.log(prediction)+(1-target)*torch.log(1-prediction))
  return loss
```

```python
#forward pass
z=w*x+b
y_pred=torch.sigmoid(z)

loss=binary_cross_entropy_loss(y_pred,y)
```

```python
loss
```

```python
#Derivatives:
#1. dL/d(y_pred) : Loss with respect to the prediction(y_pred)
dloss_dy_pred = (y_pred-y)/(y_pred*(1-y_pred))

#2. dy_pred/dz: Prediction(y_pred) with respect to z (sigmoid derivative)
dy_pred_dz = y_pred*(1-y_pred)

#3. dz/dw and dz/db : z with respect to w and b
dz_dw = x
dz_db = 1

dL_dw = dloss_dy_pred * dy_pred_dz * dz_dw
dL_db = dloss_dy_pred * dy_pred_dz * dz_db
```

```python
print(f"Manual Gradient of loss w.r.t weight(dw): {dL_dw}")
print(f"Manual Gradient of loss w.r.t bias(db): {dL_db}")
```
